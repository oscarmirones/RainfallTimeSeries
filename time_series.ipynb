```python
time = pd.date_range('1998-01-01', '2019-08-31', freq = '3H')
fechas = time.strftime('%Y%m%d-%H.nc4')
```

```python
precips = []
idx = time
for i in range(0,len(time)):
    try:
        data = xr.open_dataset(fechas[i]).precipitation[0,251,330]
        precips.append(data)
    except:
        idx = idx.drop(idx[i])
        pass
    
    
#np.where(precips == np.max(np.nan_to_num(precips, nan = 0)))
precips[50861] = 0

s = pd.Series(precips, index=idx)
```

```python
idx.to_pydatetime()
```

```python
import matplotlib.dates as dates
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
fig, ax = plt.subplots()
ax.plot_date(idx.to_pydatetime(), s, '-')
ax.xaxis.set_major_locator(dates.YearLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%Y'))
plt.tight_layout()
plt.title("San Diego's Rainfall 1998-2019")
plt.show()
#hay datos desde 1998-01-01 hasta 2019-09-01
```
